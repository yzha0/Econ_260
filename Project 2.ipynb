## II Model

The demand curve is one of the most important things in this project. In this project, I assume that both cafés only produce the identical good, braised pork rice, because it’s the only product that is identical in our cafés and it’s one of the hottest sell goods. I modeled the demand curve as below. 

\begin{align}
Q_{D(S,p,p_s,Y)} = S - ap + bp_s + cY
\end{align}

Where $S$ stands for the population of students within the school, $p$ stands for the price of the good, braised pork rice, $p_s$ stand for the price of the substitute good, which is the meal in our dining hall, and $Y$ stand for the income, or the budget that their family provide for meal in school. $a$ and $b$ are the coefficient that marks the consumer's preferance toward their meal.

The production function of two cafés is defined as below:
\begin{align}
q_{1(L,K)} = L^{\alpha_1}K^{\beta_1}\\
q_{2(L,K)} = L^{\alpha_2}K^{\beta_2}
\end{align}
Where L stand for the number of labors, and K stand for the unit of capital. I assumed that these two firms as the exact same form productive function, since their works are nearly identical. $\alpha$ and $\beta$ stands for the coefficients that may divergent from each cafe, which is determinant by the different cooker they used ($\beta$), different stimulate policy of cafe to the worker ($\alpha), and other factors that may affect their production rate through affecting the use of labor and capital. 

The cost of the café is defined below:
\begin{align}
C_{1(L,K)} = w_1L+r_1K + F_1\\
C_{2(L,K)} = w_2L+r_2K + F_2
\end{align}
Where $w$ stands for the wage rate for the labor, and $r$ stands for the interest rate for the capital, and $F$ stands for fixed cost. I assume we are in short term, and two cafes have a divergent of $w$ and $r$, because different companies may have different policies toward their labors. 


Note: 
list of general assumptions for models:
1. There are only two cafes within the school;
2. Two cafes are work in short term; 
3. They all have the Cobb-douglas production function;
4. Student only can choose between foods in cafes and dinning hall; 
5. Cafe have no control of price, price is only determined by demand and supply. 

## III Analysis

### Solve the model

For individual café, we need to find the most efficient combination of their labor and capital. After that, we can find the production function that we may use in the Cournot model. The first thing that we need to do, is set up the cost minimization problem for individual café. Note that sicne the production function is highly similar, so I use $\alpha$ and $\beta$ instead of $\alpha_1,\alpha_2$ and $\beta_1, \beta_2$, $w$, $F$ and $r$ instead of $w_1,w_2$, $r_1, r_2$ ,and $F_1, F_2$though my calculation. However, I will distinct these variables later. 

\begin{align}
&\min_{L,K} \quad C_{(L,K)} = wL+rK+F\\
\text{s.t.}\quad & q_{(L,K)} = L^{\alpha}K^{\beta}
\end{align}

Then we write the Lagrangian of the affiliate to the functions above is below:
\begin{align}
\mathcal{L} = wL+rK+F+\lambda[q-q_{L,K}] = wL+rK+F+\lambda[q-L^\alpha K^\beta]
\end{align}
Therefore the FOCs are below:
\begin{align}
\frac{\partial}{\partial L}\mathcal{L} &=  w - \frac{K^\alpha L^\beta \beta \lambda }{L} = 0 & (1.1)\\
\frac{\partial}{\partial K}\mathcal{L} &= r - \frac{K^\alpha L^\beta \alpha \lambda }{K} = 0 &(1.2)\\
\frac{\partial}{\partial \lambda}\mathcal{L} &= \overline{q} - K^\alpha L^\beta = 0 &(1.3)
\end{align}

Then we can solve for MRTS:
\begin{align}
\frac{w}{r} &= \frac{K^\alpha L^\beta \beta \lambda }{L}\frac{K}{K^\alpha L^\beta \alpha \lambda}\\
K &= \frac{wL\alpha}{r\beta} &(1.4)
\end{align}  

After plug K into (1.3), The optimal $L^*$ is:  
\begin{align}
\overline{q} &= L^{\alpha + \beta} \frac{w \alpha}{r \beta}^{\alpha } \\
L^* &= (\overline{q} (\frac{r\beta}{w\alpha})^{\alpha})^\frac{1}{\alpha + \beta}\\
&= \overline{q}^{\frac{1}{\alpha+\beta}} (\frac{r\beta}{w\alpha})^\frac{\alpha}{\alpha + \beta}
\end{align}  

Then we plug $L^*$ into (1.4), the optimal $K^*$ is:  
\begin{align}
K^* = \overline{q}^{\frac{1}{\alpha + \beta}}  (\frac {w\alpha}{r\beta})^{\frac{\beta}{\alpha + \beta}}
\end{align}

Plug $L^*$ and $K^*$ we derived from the last question into $C =wL+rK$, so we can get the long term cost function.  
\begin{align}
C_{(L,K)}&=wL+rK+F\\
&= w (\bar{q})^{\frac{1}{\alpha+\beta}}(\frac{r\beta}{w\alpha})^{\frac{\alpha}{\alpha+\beta}}+r(\bar{q})^{\frac{1}{\alpha+\beta}} (\frac{w\alpha}{r\beta})^{\frac{\beta}{\alpha+\beta}}+F\\
&= \bar{q}^{\frac{1}{\alpha+\beta}}(w^{(1-\frac{\alpha}{\alpha+\beta})} (\frac{r\beta}{\alpha})^{\frac{\alpha}{\alpha+\beta}}+r^{(1-\frac{\beta}{\alpha+\beta})}(\frac{w\alpha}{\beta})^{\frac{\beta}{\alpha+\beta}})+F\\
&=\bar{q}^{\frac{1}{\alpha+\beta}}w^{\frac{\beta}{\alpha+\beta}}((\frac{r\beta}{\alpha})^{\frac{\alpha}{\alpha+\beta}}+r^{(1-\frac{\beta}{\alpha+\beta})}(\frac{\alpha}{\beta})^{\frac{\beta}{\alpha+\beta}})+F\\
&=\bar{q}^{\frac{1}{\alpha+\beta}}w^{\frac{\beta}{\alpha+\beta}}r^{\frac{\alpha}{\alpha+\beta}}((\frac{\beta}{\alpha})^{\frac{\alpha}{\alpha+\beta}}+(\frac{\alpha}{\beta})^{\frac{\beta}{\alpha+\beta}})+F\\
&=\bar{q}^{\frac{1}{\alpha+\beta}}w^{\frac{\beta}{\alpha+\beta}}r^{\frac{\alpha}{\alpha+\beta}}(\alpha+\beta)(\frac{1}{\alpha})^{\frac{\alpha}{\alpha+\beta}}(\frac{1}{\beta})^{\frac{\beta}{\alpha+\beta}}+F\\
\end{align}

As a result, the equations below are the cost functions for each cafe:
\begin{align}
C_{1(L,K)} &= \bar{q_1}^{\frac{1}{\alpha_1+\beta_1}}w^{\frac{\beta_1}{\alpha_1+\beta_1}}r^{\frac{\alpha_1}{\alpha_1+\beta_1}}(\alpha_1+\beta_1)(\frac{1}{\alpha_1})^{\frac{\alpha_1}{\alpha_1+\beta_1}}(\frac{1}{\beta_1})^{\frac{\beta_1}{\alpha_1+\beta_1}}+F\\
C_{2(L,K)} &= \bar{q_2}^{\frac{1}{\alpha_2+\beta_2}}w^{\frac{\beta_2}{\alpha_2+\beta_2}}r^{\frac{\alpha_2}{\alpha_2+\beta_2}}(\alpha_2+\beta_2)(\frac{1}{\alpha_2})^{\frac{\alpha_2}{\alpha_2+\beta_2}}(\frac{1}{\beta_2})^{\frac{\beta_2}{\alpha_2+\beta_2}}+F
\end{align}

This outcome is long; however, this result is expected. Throughout the whole equation, where $w^{\frac{\beta_1}{\alpha_1+\beta_1}}r^{\frac{\alpha_1}{\alpha_1+\beta_1}}(\alpha_1+\beta_1)(\frac{1}{\alpha_1})^{\frac{\alpha_1}{\alpha_1+\beta_1}}(\frac{1}{\beta_1})^{\frac{\beta_1}{\alpha_1+\beta_1}}$ in the cost function of cafe 1 and $w^{\frac{\beta_2}{\alpha_2+\beta_2}}r^{\frac{\alpha_2}{\alpha_2+\beta_2}}(\alpha_2+\beta_2)(\frac{1}{\alpha_2})^{\frac{\alpha_2}{\alpha_2+\beta_2}}(\frac{1}{\beta_2})^{\frac{\beta_2}{\alpha_2+\beta_2}}$ in the cost function of cafe 2 are just constant that determined by $w, r, \alpha, \beta$. 

For convience, here we let $w^{\frac{\beta_1}{\alpha_1+\beta_1}}r^{\frac{\alpha_1}{\alpha_1+\beta_1}}(\alpha_1+\beta_1)(\frac{1}{\alpha_1})^{\frac{\alpha_1}{\alpha_1+\beta_1}}(\frac{1}{\beta_1})^{\frac{\beta_1}{\alpha_1+\beta_1}} = A_1$ and $w^{\frac{\beta_2}{\alpha_2+\beta_2}}r^{\frac{\alpha_2}{\alpha_2+\beta_2}}(\alpha_2+\beta_2)(\frac{1}{\alpha_2})^{\frac{\alpha_2}{\alpha_2+\beta_2}}(\frac{1}{\beta_2})^{\frac{\beta_2}{\alpha_2+\beta_2}} = A_2$. Then we can rewrite the cost function in a simplifed way:
\begin{align}
C_{1(q_1)} &= A_1\bar{q_1}^{\frac{1}{\alpha_1+\beta_1}}+F&(2.1)\\
C_{2(q_2)} &= A_2\bar{q_2}^{\frac{1}{\alpha_2+\beta_2}}+F&(2.2)
\end{align}

By this model, it's easier for us to discuss how does the output of cafe may looks like if we increase the scale. We approach it by using derivatives, and analyse the cost for $\overline{q}$ increase to infinity. We take cafe 1 as an example, since we have the same form of cost function.
\begin{align}
\frac{d}{dq} C_{1(L,K)} = \frac{d}{dq} (A\bar{q_1}^{\frac{1}{\alpha_1+\beta_1}}+F) = \frac{A}{\alpha_1+\beta_1} \bar{q_1}^{\frac{1}{\alpha_1+\beta_1}-1}
\end{align}

If $\alpha + \beta > 1$, $ \frac{1}{\alpha +\beta} < 1$, so when $\bar{q}$ goes to infinity, the marginal cost will be 0, so the production function follows economies of scale.\
If $\alpha + \beta < 1$, $ \frac{1}{\alpha +\beta} > 1$, so when $\bar{q}$ goes to infinity, the marginal cost will be infinity, so the production function follows diseconomies of scale.\
If $\alpha + \beta = 1$, $ \frac{1}{\alpha +\beta} = 1$, so when $\bar{q}$ goes to infinity, the marginal cost will hold constant, so the production function follows constant return to scale.

Now we have the cost function for individual cafe, but we still need the demand function for the entire market at school. We write the demand function in the model part into the inversed demand function below:
\begin{align}
Q_{D(S,p,p_s,Y)} = S - ap + bp_s + cY\\
p_{(S,p_s,Y,Q_D)}= \frac{1}{a}S + \frac{b}{a}p_s + \frac{c}{a}Y -\frac{1}{a} Q_D
\end{align}

Since $Q_D$ is conposed by only two cafes, we can substitute $Q_D$ as $q_1+q_2$. As a result, the inverse demand wil be:

$$p_{(S,p_s,Y,q_1,q_2)}= \frac{1}{a}S + \frac{b}{a}p_s + \frac{c}{a}Y -\frac{1}{a} (q_1+q_2)$$

We can see that in the inverse demand function above, $\frac{1}{a}S + \frac{b}{a}p_s + \frac{c}{a}Y$ is a constant. So in order to simplify the calculation process, let $D = \frac{1}{a}S + \frac{b}{a}p_s + \frac{c}{a}Y$. Then the function will be:
\begin{align}
p_{(C, q_1,q_2)}= D -\frac{1}{a}(q_1+q_2)
\end{align}

After determined the demand function for the market, and the cost function of cafes, we are ready to find how the firms may behave by using the Cournet model. In this model we assume that rival cafes produce a homogenous product, and each attempts to maximize profits by choosing how much to produce. All firms choose output (quantity) simultaneously.

We first write down the cafes' profit macimization problems. 

For cafe 1, we have:
\begin{align}
\max_{q_1}\pi_1 &= p(D,q_1,q_2)q_1 -C_1(q_1)\\
\max_{q_1}\pi_1 &=[D -\frac{1}{a}(q_1+q_2)]q_1-(A_1q_1^{\frac{1}{\alpha_1+\beta_1}}+F)\\
&= -\frac{1}{a}q_1^2 - \frac{1}{a}q_1q_2 - A_1 q_1^{\frac{1}{\alpha_1+\beta_1}}-F+Dq_1
\end{align}

For cafe 2, we have:
\begin{align}
\max_{q_2}\pi_2 &= p(D,q_1,q_2)q_2 -C_2(q_2)\\
\max_{q_2}\pi_2 &=[D -\frac{1}{a}(q_1+q_2)]q_2-(A_2q_2^{\frac{1}{\alpha_2+\beta_2}}+F)\\
&= -\frac{1}{a}q_2^2 - \frac{1}{a}q_1q_2 - A_2 q_2^{\frac{1}{\alpha_2+\beta_2}}-F +Dq_2
\end{align}

Now, we derive the following First Order Conditions (FOC's) for each cafe. We do so by taking the partial derivatives of our profit functions with respect to $q_1$ and $q_2$ and set them equals to 0.
\begin{align}
\frac{\partial \pi_1}{\partial q_1} = -\frac{2}{a}q_1-\frac{1}{a}\hat{q_2} - \frac{A_1}{\alpha_1+\beta_1}q_1^{\frac{1}{\alpha_1+\beta_1}-1}+D =0\\
\frac{\partial \pi_2}{\partial q_2} = -\frac{2}{a}q_2-\frac{1}{a}\hat{q_1} - \frac{A_2}{\alpha_2+\beta_2}q_2^{\frac{1}{\alpha_2+\beta_2}-1}+D =0\\
\end{align}

From the equations above, we can solve for the optimal outputs $q_1^*$ and $q_2^*$. However, this equation is really difficult to solve, so I assume that both cafe have diseconomics of scale, and $\alpha+\beta = 0.5$.

Then we can easily get the optimal output as below:
\begin{align}
q_1^* = \frac{-\frac{1}{a}\hat{q_2}+D}{\frac{2}{a}+2A_1}\\
q_2^* = \frac{-\frac{1}{a}\hat{q_1}+D}{\frac{2}{a}+2A_2}
\end{align}

Therefore, we can find the Nash Equilibrium best response by assuming the believe for each cafe is true, which means $\hat{q_1}=q_1^*$ and $\hat{q_2}=q_2^*$.

\begin{align}
q_1^* = \frac{-\frac{1}{a}q_2^*+D}{\frac{2}{a}+2A_1}\\
q_2^* = \frac{-\frac{1}{a}q_1^*+D}{\frac{2}{a}+2A_2}
\end{align}

After we substitute $q_1^*$ into $q_2^*$, and $q_2^*$ into $q_1^*$, we can get the equilibrium below:

$$q_1^* = \frac{\frac{D}{2A_1+\frac{2}{a}}-\frac{D}{a(2A_2+\frac{2}{a})}}{1-\frac{1}{a^2(2A_2+\frac{2}{a})(2A_1+\frac{2}{a})}}$$

$$q_2^* = \frac{\frac{D}{2A_2+\frac{2}{a}}-\frac{D}{a(2A_1+\frac{2}{a})}}{1-\frac{1}{a^2(2A_1+\frac{2}{a})(2A_2+\frac{2}{a})}}$$

\max_{q_2}\pi_2 = -\frac{1}{a}q_2^{*2} - \frac{1}{a}q_1^*q_2^* - A_2 q_2^{*\frac{1}{\alpha_2+\beta_2}}-F +Dq_2^*
